## Linear Regression using Scikit-learn

### 1. Prepare the Data

```python
# Download data boston housing dataset
```

```python
import sklearn
sklearn.__version__
```

```python
!pip install --upgrade scikit-learn==1.0.2
```

```python
#split data into train and test
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
print(data['DESCR'])
```

```python
features = data['data']
target = data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = features
y = target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### 2. Build the model and train it

```python
#train model
from sklearn.linear_model import LinearRegression
```

```python
model_linreg = LinearRegression()
model_linreg.fit(X_train, y_train)
```

### 3. Let's Evaluate our model using mean squared errorr, mean absolute error, and r2 score

```python
#evaluate model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from math import sqrt
```

```python
y_test_pred = model_linreg.predict(X_test)
```

```python
print(f"MAE: {mean_absolute_error(y_test, y_test_pred)}")
print(f"MSE: {mean_squared_error(y_test, y_test_pred)}")
print(f"RMSE: {sqrt(mean_squared_error(y_test, y_test_pred))}")
print(f"R^2: {r2_score(y_test, y_test_pred)}")
```

## Logistic Regression using Scikit-Learn

### Binary

#### 1. Prepare the data

```python
# download data iris dataset
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
print(data['DESCR'])
```

```python
features = data['data']
target = data['target']
```

```python
target
```

```python
len(target[target<2])
```

```python
# setosa vs versicolor (0 vs 1) -> binary classification
target_binary = target[target<2]
features_binary = features[:100]
```

```python
# split data into train and test
X = features_binary
y = target_binary
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

#### 2. Build the model and train it

```python
# train model
from sklearn.linear_model import LogisticRegression
```

```python
model_logreg = LogisticRegression(random_state=1)
model_logreg.fit(X_train, y_train)
```

#### 3. Evaluate the model

```python
# evaluate model
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, recall_score, f1_score, classification_report
```

```python
y_test_pred = model_logreg.predict(X_test)
```

```python
import pandas as pd
pd.Series(y_test).value_counts()
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_test_pred)}")
print(f"Precision: {precision_score(y_test, y_test_pred)}")
print(f"Recall: {recall_score(y_test, y_test_pred)}")
print(f"F1 score: {f1_score(y_test, y_test_pred)}")
print(f"Classification report:\n{classification_report(y_test, y_test_pred)}")
```

### Multiclass

#### 1. Prepare the data

```python
# setosa vs versicolor vs virginica (0 vs 1 vs 2) -> multiclass classification
```

```python
# split data into train and test
X = features
y = target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
pd.Series(y_train).value_counts()
```

```python
pd.Series(y_test).value_counts()
```

#### 2. Build the model and train it

```python
model_logreg = LogisticRegression(random_state=1)
model_logreg.fit(X_train, y_train)
```

#### 3. evaluate the model

```python
y_test_pred = model_logreg.predict(X_test)
```

##### Performance on Test set

```python
print(f"Accuracy: {accuracy_score(y_test, y_test_pred)}")
print(f"Precision: {precision_score(y_test, y_test_pred, average='weighted')}")
print(f"Recall: {recall_score(y_test, y_test_pred, average='weighted')}")
print(f"F1 score: {f1_score(y_test, y_test_pred, average='weighted')}")
print(f"Classification report:\n{classification_report(y_test, y_test_pred)}")
```

##### Performance on Train set

```python
y_train_pred = model_logreg.predict(X_train)
```

```python
print(f"Accuracy: {accuracy_score(y_train, y_train_pred)}")
print(f"Precision: {precision_score(y_train, y_train_pred, average='weighted')}")
print(f"Recall: {recall_score(y_train, y_train_pred, average='weighted')}")
print(f"F1 score: {f1_score(y_train, y_train_pred, average='weighted')}")
```

- performa train sangat jauh nilainya dibandingkan test (train > test) -> overfitting di data training
- performa di kedua set, train dan test itu sangat kecil -> underfitting
- performa train < test -> unusual (a sign of data leakage and other problems during data preprocessing)
