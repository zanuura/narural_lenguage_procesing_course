# Get data from a single page

## Base code

```python
import requests
from bs4 import BeautifulSoup

# url = 'https://www.detik.com/tag/pendidikan'
base_url = 'https://www.detik.com/search/searchall?query='
query = 'pendidikan'

# Mengirim permintaan GET ke URL
response = requests.get(base_url + query)

# Mengecek status kode respons
if response.status_code == 200:
    # Mengubah respons menjadi objek BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
    berita_elements = soup.select('.content .list-berita article')

    # Mengiterasi setiap elemen berita
    for berita in berita_elements:
        # Mendapatkan judul dan URL berita
        judul = berita.select_one('.box_text .title').text.strip()
        url_berita = berita.select_one('a')['href']
        summary_berita = berita.select_one('p').text.strip()

        print('Judul:', judul)
        print('URL Berita:', url_berita)
        print('Summary Berita:', summary_berita)
        print('---')
else:
    print('Gagal mengambil data. Status kode:', response.status_code)
```

## Save the data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_single_page(base_url, query):

  # Mengirim permintaan GET ke URL
  response = requests.get(base_url + query)

  # Mengecek status kode respons
  if response.status_code == 200:
      # Mengubah respons menjadi objek BeautifulSoup
      soup = BeautifulSoup(response.text, 'html.parser')

      # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
      berita_elements = soup.select('.content .list-berita article')

      # Mengiterasi setiap elemen berita
      list_judul = []
      list_url = []
      list_summary = []
      for berita in berita_elements:
          # Mendapatkan judul dan URL berita
          judul = berita.select_one('.box_text .title').text.strip()
          url_berita = berita.select_one('a')['href']
          summary_berita = berita.select_one('p').text.strip()

          list_judul.append(judul)
          list_url.append(url_berita)
          list_summary.append(summary_berita)
  else:
      pass

  data = pd.DataFrame({
      'judul':list_judul,
      'url':list_url,
      'summary':list_summary
  })

  return data

```

# Get data from multiple pages

## Base code

```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://www.detik.com/search/searchall?'
query = 'pendidikan'
sort_by = 'time'
page_count = 1

for page in range(1, page_count+1):
    url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

    # Mengirim permintaan GET ke URL
    response = requests.get(url)
    print('Page:', page)
    # Mengecek status kode respons
    if response.status_code == 200:
        # Mengubah respons menjadi objek BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
        berita_elements = soup.select('.content .list-berita article')

        # Mengiterasi setiap elemen berita
        for berita in berita_elements:
            # Mendapatkan judul dan URL berita
            judul = berita.select_one('.box_text .title').text.strip()
            url_berita = berita.select_one('a')['href']
            category = berita.select_one('span .date .category').text.strip()
            date = berita.select_one('span .date .category').find_next_sibling(string=True)
            summary_berita = berita.select_one('p').text.strip()

            print('Judul:', judul)
            print('URL Berita:', url_berita)
            print('Kategori:', category)
            print('Waktu Berita:', date)
            print('Summary Berita:', summary_berita)
            print('---')
    else:
        print('Gagal mengambil data. Status kode:', response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://www.detik.com/search/searchall?'
query = 'pendidikan'
sort_by = 'time'
page_count = 1

for page in range(1, page_count+1):
    url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

    # Mengirim permintaan GET ke URL
    response = requests.get(url)
    print('Page:', page)
    # Mengecek status kode respons
    if response.status_code == 200:
        # Mengubah respons menjadi objek BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
        berita_elements = soup.select('.content .list-berita article')

        # Mengiterasi setiap elemen berita
        for berita in berita_elements:
            # Mendapatkan judul dan URL berita
            judul = berita.select_one('.box_text .title').text.strip()
            url_berita = berita.select_one('a')['href']
            category = berita.select_one('span .date .category').text.strip()
            date = berita.select_one('span .date .category').find_next_sibling(string=True)
            summary_berita = berita.select_one('p').text.strip()

            print('Judul:', judul)
            print('URL Berita:', url_berita)
            print('Kategori:', category)
            print('Waktu Berita:', date)
            print('Summary Berita:', summary_berita)

            response_detail = requests.get(url_berita)
            # Mengecek status kode respons halaman detail
            if response_detail.status_code == 200:
                soup_detail = BeautifulSoup(response_detail.text, 'html.parser')

                # Mencari elemen detail
                detail_berita = ' '.join([i.text.strip() for i in soup_detail.select_one('.itp_bodycontent').find_all('p')])
                author = soup_detail.find('meta', attrs={'name': 'dtk:author'})['content']

                print('Detail:', detail_berita)
                print('Penulis:', author)


            else:
                print('Gagal mengambil halaman detail. Status kode:', response.status_code)

            print('---')
    else:
        print('Gagal mengambil data. Status kode:', response.status_code)
```

## Save the data

```python
import requests
from bs4 import BeautifulSoup

def get_multiple_pages(base_url, query, sort_by, page_count, get_detail=True):

    list_judul = []
    list_url = []
    list_kategori = []
    list_date = []
    list_summary = []
    list_detail_berita = []
    list_author = []

    for page in range(1, page_count+1):
        url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

        # Mengirim permintaan GET ke URL
        response = requests.get(url)
        # Mengecek status kode respons
        if response.status_code == 200:
            # Mengubah respons menjadi objek BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Mencari elemen dengan class 'list-berita' yang berada di dalam div dengan class 'content'
            berita_elements = soup.select('.content .list-berita article')

            # Mengiterasi setiap elemen berita
            for berita in berita_elements:
                # Mendapatkan judul dan URL berita
                judul = berita.select_one('.box_text .title').text.strip()
                url_berita = berita.select_one('a')['href']
                category = berita.select_one('span .date .category').text.strip()
                date = berita.select_one('span .date .category').find_next_sibling(string=True)
                summary_berita = berita.select_one('p').text.strip()

                list_judul.append(judul)
                list_url.append(url_berita)
                list_kategori.append(category)
                list_date.append(date)
                list_summary.append(summary_berita)

                if get_detail:
                    response_detail = requests.get(url_berita)
                    # Mengecek status kode respons halaman detail
                    if response_detail.status_code == 200:
                        soup_detail = BeautifulSoup(response_detail.text, 'html.parser')

                        # Mencari elemen detail
                        detail_berita = ' '.join([i.text.strip() for i in soup_detail.select_one('.itp_bodycontent').find_all('p')])
                        author = soup_detail.find('meta', attrs={'name': 'dtk:author'})['content']

                        list_detail_berita.append(detail_berita)
                        list_author.append(author)

        else:
            pass

    data = pd.DataFrame({
        'judul':list_judul,
        'url':list_url,
        'kategori':list_kategori,
        'tanggal':list_date,
        'summary':list_summary,
        'isi_berita':list_detail_berita,
        'penulis':list_author,

    })

    return data
```

```python
df_detail = get_multiple_pages(base_url='https://www.detik.com/search/searchall?', query='pendidikan', sort_by='time', page_count=2, get_detail=True)
display(df_detail.info())
display(df_detail.head())
```
