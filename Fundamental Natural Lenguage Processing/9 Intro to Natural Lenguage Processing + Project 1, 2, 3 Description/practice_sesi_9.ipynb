```python
import requests 
from bs4 import BeautifulSoup

base_url = 'https://www.detik.com/search/searchall?query='
query = 'pendidikan'

response = requests.get(base_url + query)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    berita_elements = soup.select('.content .list-berita article')

    for berita in berita_elements:

        judul = berita.select_one('.box_text .title').text.strip()
        url_berita = berita.select_one('a')['href']
        summery_berita = berita.select_one('p').text.strip()

        print('Judul:', judul)
        print('URL Berita:', url_berita)
        print('Summery Berita:', summery_berita)
        print('--------------------------------')

else:
    print('Gagal mengambil data. Status kode:', response.status_code)
```

## Save The Data

```python
import pandas as pd

def get_single_page(base_url, query):

    response = requests.get(base_url + query)

    if response.status_code == 200:
         soup = BeautifulSoup(response.text, 'html.parser')

         berita_elements = soup.select('.content .list-berita article')

         list_judul = []
         list_url = []
         list_summery = []

         for berita in berita_elements:
              
              judul = berita.select_one('.box_text .title').text.strip()
              url_berita = berita.select_one('a')['href']
              summery_berita = berita.select_one('p').text.strip()

              list_judul.append(judul)
              list_url.append(url_berita)
              list_summery.append(summery_berita)

    else:
         pass
    
    data = pd.DataFrame({
         'judul': list_judul,
         'url': list_url,
         'summery': list_summery,
    })

    return data
```

## Save and get Data from Multiple Page

```python
def get_multiple_pages(base_url, query, sort_by, page_count):

    list_judul = []
    list_url = []
    list_kategori = []
    list_date = []
    list_summery = []

    for page in range(1, page_count+1):
        url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            berita_elements = soup.select('.content .list-berita article')

            for berita in berita_elements:
                judul = berita.select_one('.box_text .title').text.strip()
                url_berita = berita.select_one('a')['href']
                category = berita.select_one('span .date .category').text.strip()
                date = berita.select_one('span .date .category').find_next_sibling(string=True)
                summery_berita = berita.select_one('p').text.strip()

                list_judul.append(judul)
                list_url.append(url_berita)
                list_kategori.append(category)
                list_date.append(date)
                list_summery.append(summery_berita)
        else:
            pass

    data = pd.DataFrame({
        'judul': list_judul,
        'url': list_url,
        'kategori': list_kategori,
        'date': list_date,
        'summery': list_summery,
    })

    return data
```

```python
df_multiple = get_multiple_pages(base_url='https://www.detik.com/search/searchall?', query='pendidikan', sort_by='time', page_count=3)
display(df_multiple.info())
display(df_multiple.head())
```

# Get Detail Content

```python
base_url = 'https://www.detik.com/search/searchall?'
query = 'pendidikan'
sort_by = 'time'
page_count = 1

for page in range(1, page_count+1):
    url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

    response = requests.get(url)
    print('Page: ', page)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        berita_elements = soup.select('.content .list-berita article')

        for berita in berita_elements:
            judul = berita.select_one('.box_text .title').text.strip()
            url_berita = berita.select_one('a')['href']
            category = berita.select_one('span .date .category').text.strip()
            date = berita.select_one('span .date .category').find_next_sibling(string=True)
            summary_berita = berita.select_one('p').text.strip()

            print('Judul:', judul)
            print('URL Berita:', url_berita)
            print('Kategori:', category)
            print('Waktu Berita:', date)
            print('Summary Berita:', summary_berita)

            response_detail = requests.get(url_berita)

            if response_detail.status_code == 200:
                soup_detail = BeautifulSoup(response_detail.text, 'html.parser')

                detail_berita = ' '.join([i.text.strip() for i in soup_detail.select_one('.itp_bodycontent').find_all('p')])
                author = soup_detail.find('meta', attrs={'name': 'dtk:author'})['content']

                print('Detail:', detail_berita)
                print('Penulis:', author)

            else:
                print('Gagal mengambil halaman detail. Status kode:', response.status_code)

            print('------')
    else:
        print('Gagal mengambil data. Status kode:', response.status_code)
```

# Save the Data

```python
def get_multiple_details_page(base_url, query, sort_by, page_count, get_detail=True):
    list_judul = []
    list_url = []
    list_kategori = []
    list_date = []
    list_summary = []
    list_detail_berita = []
    list_author = []

    for page in range(1, page_count+1):
        url = base_url + '&query=' + query + '&sortby=' + sort_by + '&page=' + str(page)

        response = requests.get(url)
        print('Page: ', page)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            berita_elements = soup.select('.content .list-berita article')

            for berita in berita_elements:
                judul = berita.select_one('.box_text .title').text.strip()
                url_berita = berita.select_one('a')['href']
                category = berita.select_one('span .date .category').text.strip()
                date = berita.select_one('span .date .category').find_next_sibling(string=True)
                summary_berita = berita.select_one('p').text.strip()

                print('Judul:', judul)
                print('URL Berita:', url_berita)
                print('Kategori:', category)
                print('Waktu Berita:', date)
                print('Summary Berita:', summary_berita)

                if get_detail:
                    response_detail = requests.get(url_berita)

                    if response_detail.status_code == 200:
                        soup_detail = BeautifulSoup(response_detail.text, 'html.parser')

                        # detail_berita = ' '.join([i.text.strip() for i in soup_detail.select_one('.itp_bodycontent').find_all('p')])
                        # author = soup_detail.find('meta', attrs={'name': 'dtk:author'})['content']

                        # list_detail_berita.append(detail_berita)
                        # list_author.append(author)
                        itp_bodycontent = soup_detail.select_one('.itp_bodycontent')
                        if itp_bodycontent:
                            paragraphs = itp_bodycontent.find('p')

                            detail_berita = ' '.join([p.text.strip() for p in paragraphs])
                            author = soup_detail.find('meta', attrs={'name': 'dtk:author'})['content']

                            list_detail_berita.append(detail_berita)
                            list_author.append(author)
                        else:
                            print('Element with class "itp_bodycontent" not found on the page.')
                        

                    else:
                        print('Gagal mengambil halaman detail. Status kode:', response.status_code)

                print('------')
        else:
            print('Gagal mengambil data. Status kode:', response.status_code)

    data = pd.DataFrame({
        'judul': list_judul,
        'url': list_url,
        'kategori': list_kategori,
        'tanggal': list_date,
        'summery': list_summary,
        'isi_berita': list_detail_berita,
        'penulis': list_author,
    })

    return data
```

```python
df_detail = get_multiple_details_page(base_url='https://www.detik.com/search/searchall?', query='pendidikan', sort_by='time', page_count=2, get_detail=True)
display(df_detail.info())
display(df_detail.head())
```
