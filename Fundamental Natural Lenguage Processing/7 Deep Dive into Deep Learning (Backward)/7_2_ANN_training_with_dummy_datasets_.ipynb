## 1. create a linear function for the dummy datasets

for this tutorial we are gonna make this function for the datasets,
f(x1,x2,x3,x4) = x1*w1+x2*w2+x3*w3+x4*w4
w1 = 1
w2 = 2
w3 = 3
w4 = 4

```python
def function(x1, x2, x3, x4):
  w1,w2,w3,w4 = 1,2,3,4
  return  x1*w1+x2*w2+x3*w3+x4*w4
```

```python
#example
x1,x2,x3,x4 = 1,2,3,4
function(x1,x2,x3,x4) #1+4+9+16=30
```

## 2. create dummy datasets


```python
import numpy as np
```

```python
def create_dummy_datasets(n=1000):
  x = np.random.randint(0, 2000, size=(n,4))

  y = []
  for i in x:
    y_temp = function(*i) #i[0], i[1], ..., i[n-1] n: panjang arraynya
    y.append(y_temp)
  
  y = np.asarray(y)

  return x,y
```

```python
N_SAMPLE = 1000
data = create_dummy_datasets(N_SAMPLE)
```

```python
print(data[0]) #for the x data
print(data[1]) #for the y data
```

```python
#lets convert it to torch.tensor dtype torch.float32 so it can run smoothly on neural nets
import torch
```

```python
X = torch.tensor(data[0], dtype = torch.float32)
y = torch.tensor(data[1], dtype = torch.float32)
```

```python
print(X)
print(y)
```

## 3.lets build a model

```python
from torch import nn
```

```python
class RegressionModel(nn.Module):
  def __init__(self, input_node=4, output_node=1):
    super().__init__()
    self.input_node = input_node
    self.output_node = output_node

    self.fc = nn.Linear(self.input_node, self.output_node, bias=False) #we want to make the bias false so it can relate with the function but in most cases you need the bias to be true
  
  def forward(self,x):
    return self.fc(x)
```

## 4. Let's Create training scheme
before that let's understand about concept of optimization and loss function

```python
sample_model = RegressionModel()
optim = torch.optim.SGD(sample_model.parameters(), lr=0.001) #Stochastic Gradient Descent
loss_fn = nn.MSELoss() #Mean Squared error loss
```

```python
sample_X = X
sample_y = y

BATCH_SIZE = 10
sample_X = torch.reshape(sample_X, (N_SAMPLE//BATCH_SIZE, BATCH_SIZE, 4)) #let's make batch 10 | (N_SAMPLE,4) -> (N_SAMPLE//BATCH_SIZE, BATCH_SIZE, 4) -> (100, 10, 4)
sample_y = torch.reshape(sample_y, (N_SAMPLE//BATCH_SIZE, BATCH_SIZE, 1)) #let's make batch 10 | (N_SAMPLE,1) -> (N_SAMPLE//BATCH_SIZE, BATCH_SIZE, 1) -> (100, 10, 1)

sample_X = sample_X/2000
sample_y = sample_y/2000
```

```python
print(sample_X.shape)
print(sample_y.shape)
print(sample_model)
```

```python
#this is what we call one epoch training
for i in range(1000):
  print("EPOCH ", i)
  for i, (inputs, target) in enumerate(zip(sample_X, sample_y)):
    output = sample_model(inputs)
    loss = loss_fn(output,target)
    if i % 10 == 0:
      print("   step ",i)
      print(loss.item())
    # Backpropagation
    optim.zero_grad()
    loss.backward()
    optim.step()
```

```python
#lets see the total loss
loss_total = []
for inputs, target in zip(sample_X, sample_y):
  output = sample_model(inputs)
  loss = loss_fn(output,target)
  loss_total.append(loss.item())

loss_average = np.mean(loss_total)
```

```python
print(loss_average)
```

```python
#lets we see the parameters, that's the weight that we trained
for param in sample_model.parameters():
  print(param.data)
```

## 5. create inference function

```python
def inference(x1,x2,x3,x4):
  inputs = torch.tensor([[x1,x2,x3,x4]], dtype=torch.float32)
  inputs = inputs/2000
  output = sample_model(inputs)
  output*=2000

  return output[0].detach().numpy()
```

```python
x1,x2,x3,x4 = 3,4,5,6
inference(x1,x2,x3,x4), function(x1,x2,x3,x4)
```
