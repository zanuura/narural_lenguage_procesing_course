## TIME SERIES

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from datetime import datetime, timedelta
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid', marker=None)

plt.tight_layout()
plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

```python
#day-month-year
#pass format from mpl_class
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate() #get current figure
date_format = mpl.dates.DateFormatter('%b, %d %Y')

plt.gca().xaxis.set_major_formatter(date_format) #get current axis


plt.tight_layout()
plt.show()
```

More about date formatting: https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

### Import dataset

```python
data = pd.read_csv('./data_date.csv')
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Price')
plt.xlabel ('Date')
plt.ylabel ('Closing Price')

plt.tight_layout()
plt.show()
```

However, it is passed as `strings`\
How do we know? Let's copy older date to the last column

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Price')
plt.xlabel ('Date')
plt.ylabel ('Closing Price')

plt.tight_layout()
plt.show()
```
