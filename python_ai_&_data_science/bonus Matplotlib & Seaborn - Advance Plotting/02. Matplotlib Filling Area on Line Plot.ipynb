## FILLING AREA ON LINE PLOT

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('./data_fill.csv')
```

```python
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

### Fill below python

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),interpolate=True, alpha=0.25)

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries <= overall_median), color='red', interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.legend()
plt.tight_layout()
plt.show()
```
