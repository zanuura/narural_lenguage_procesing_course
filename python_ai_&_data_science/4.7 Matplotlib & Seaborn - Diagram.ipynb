```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat Scatter Plot

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)

x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7, 14, 12])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])

plt.scatter(x, y)

plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
size = np.array([10, 25, 50, 100, 250, 500, 30, 45, 5, 150, 300, 100, 40])

plt.scatter(x, y, s=size, color='#4caf50')
plt.show()
```

### Membuat Bar Chart

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
color = np.array(['red', 'blue', 'green', 'yellow'])

plt.bar(x, y, color=color)
plt.show()
```

### Membuat Histogram

```python
x = np.random.normal(170, 10, 250)
```

```python
plt.hist(x)
plt.show()
```

### Membuat Pie Chart

```python
colors = ['cyan', 'hotpink', 'b', '#4caf50']
x = np.array([35, 25, 25, 15])

plt.pie(x, colors=colors)
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['red', 'yellow', 'hotpink', '#4caf50']
explode = [0, 0, 0.2, 0]
x = np.array([35, 25, 25, 15])

plt.pie(x, colors=colors, labels=labels, explode=explode)
plt.legend(loc='upper left')
plt.show()
```
